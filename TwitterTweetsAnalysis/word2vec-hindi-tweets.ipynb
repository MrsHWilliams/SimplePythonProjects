```python
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize 
from sqlalchemy import create_engine
import pymysql
import pandas as pd
from gensim.models import Word2Vec, KeyedVectors


def stop_words(text):
    global hin
    word_tokens = word_tokenize(text)
    filtered_sentence = [w for w in word_tokens if not w in hin] 
    return filtered_sentence
```

```python
db_connection_str = 'mysql+pymysql://suleman:root@localhost/twitter'
db_connection = create_engine(db_connection_str)

df = pd.read_sql('SELECT tweet FROM `tweets` where lang = "hi"', con=db_connection)
```

```python
df.info()
```

```python
df["tweet"] = df["tweet"].apply(stop_words)


tweets = df["tweet"].values
```
